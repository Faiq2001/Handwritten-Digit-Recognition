```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train=pd.read_csv('mnist_train.csv')
df_test=pd.read_csv('mnist_test.csv')
```

```python
x=df_train.values[:,1:]
y=df_train.values[:,0]
x_test=df_test.values[:,1:]
y_test=df_test.values[:,0]
```

```python
def drawImage(sample):
    img=sample.reshape((28,28))
    plt.imshow(img,cmap='gray')
drawImage(x[0])
y[0]
```

```python
def distance(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(x,y,querypoint,k=5):
    vals=[]
    for i in range(x.shape[0]):
        d=distance(querypoint,x[i])
        vals.append((d,y[i]))
    vals=sorted(vals)
    vals=vals[:k]
    
    vals=np.array(vals)
    
    new_values=np.unique(vals[:,1],return_counts=True)
    print(new_values)
    
    index=new_values[1].argmax()
    pred=new_values[0][index]
    
    return pred
```

```python
pred=knn(x,y,x_test[1])
print(pred)
```

```python
drawImage(x_test[1])
```

```python
def get_acc(x,y,x_test,y_test):
    y_pred=np.zeros(x.shape[0])
    for i in range(x.shape[0]):
        y_pred[i]=knn(x,y,x_test[i])
    return float(sum(y_pred==y_test))/y_test.shape
```

```python
acc=get_acc(x,y,x_test,y_test)
plt.plot(acc)
plt.ylabel('accuracy')
plt.xlabel('time')
plt.show()
```
